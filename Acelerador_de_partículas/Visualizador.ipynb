```python
# Convertidor del archivo
import os
nombre_archivo_scad = "/workspaces/First-repository/Acelerador_de_partículas/ProyRiel3BobinaSensor.scad"
nombre_archivo_stl = "/workspaces/First-repository/Acelerador_de_partículas/ProyRiel3Bobina.stl"
def convertir_scad_a_stl(nombre_archivo_scad, nombre_archivo_stl):
  """Convierte un archivo SCAD a STL usando OpenSCAD.

  Args:
    nombre_archivo_scad: El nombre del archivo SCAD.
    nombre_archivo_stl: El nombre del archivo STL de salida.
  """
  os.system(f"openscad -o {nombre_archivo_stl} {nombre_archivo_scad}")
# Llamar a la función   
#!chmod +r /workspaces/First-repository/Acelerador_de_partículas/ProyRiel3BobinaSensor.scad
convertir_scad_a_stl(nombre_archivo_scad , nombre_archivo_stl)
```
